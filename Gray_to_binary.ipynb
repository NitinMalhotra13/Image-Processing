```python
from PIL import Image
import matplotlib.pyplot as plt

print("Place your image in the same folder as this script.")
file_name = input("Enter image file name (with extension): ")
img = Image.open(file_name).convert("L")
width, height = img.size
pixels = img.load()

print("1. Use mean intensity as threshold")
print("2. Enter threshold manually")
choice = int(input("Enter your choice (1 or 2): "))

if choice == 1:
    total = 0
    for y in range(height):
        for x in range(width):
            total += pixels[x, y]
    threshold = total // (width * height)
elif choice == 2:
    threshold = int(input("Enter threshold value (0–255): "))
else:
    print("Invalid choice.")
    exit()

print("Threshold value:", threshold)

binary_img = Image.new('1', (width, height))
for y in range(height):
    for x in range(width):
        val = 1 if pixels[x, y] > threshold else 0
        binary_img.putpixel((x, y), val)

plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(binary_img, cmap='gray')
plt.title("Binary Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```
